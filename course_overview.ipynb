# Course Welcome and Overview

Welcome to the Python Programming Course! This course is designed to introduce you to the fundamentals of programming using Python, focusing on computational thinking and problem-solving. It's a **self-paced**, **interactive**, and **open-source** learning experience. You'll engage with practical exercises, mini-projects, and unit tests directly within these Jupyter notebooks, guided by the excellent 'Think Python 2e' textbook and the 'Python Data Science Handbook'. Let's get started on your journey to thinking like a computer scientist!

## Course Modules Overview

Our course is divided into four main modules, each building upon the last to solidify your understanding of Python programming and computational thinking. Below is an overview of each module and links to the weekly notebooks.

### Module 1: The Foundation (Weeks 1–4)

This module covers the absolute basics of Python programming, including syntax, variables, and control flow, laying a strong foundation for future topics.

*   **Week 1:** [The Environment & 'Hello World'](week_01.ipynb)
*   **Week 2:** [Variables, Expressions, and Statements](week_02.ipynb)
*   **Week 3:** [Logic and Conditionals](week_03.ipynb)
*   **Week 4:** [Iteration (Loops)](week_04.ipynb)

### Module 2: Data Structures (Weeks 5–9)

Here, you'll learn about Python's fundamental data structures, which are essential for storing, organizing, and manipulating data effectively.

*   **Week 5:** [Strings](week_05.ipynb)
*   **Week 6:** [Lists](week_06.ipynb)
*   **Week 7:** [Dictionaries](week_07.ipynb)
*   **Week 8:** [Tuples](week_08.ipynb)
*   **Week 9:** [Files](week_09.ipynb)

### Module 3: Modular Programming (Weeks 10–12)

This module focuses on writing reusable code, breaking down complex problems into smaller, manageable functions, and using external libraries.

*   **Week 10:** [Functions](week_10.ipynb)
*   **Week 11:** [Debugging & Error Handling](week_11.ipynb)
*   **Week 12:** [Modules and Packages](week_12.ipynb)

### Module 4: The Capstone (Weeks 13–14)

The final module applies your learned skills to real-world data challenges using popular Data Science libraries, preparing you for more advanced topics.

*   **Week 13:** [Data Science with Pandas (Part 1)](week_13.ipynb)
*   **Week 14:** [Data Visualization with Matplotlib](week_14.ipynb)
